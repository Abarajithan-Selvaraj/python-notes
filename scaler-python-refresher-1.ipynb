```python
# print function
print('hello')
print("world")

# with more than one arguments
print("hey", 'there you', 1+1)             # comma puts a space in between (space is the default seperator)
print("hey", 'there you', 1+1, sep='_')    # sep='_': puts an underscore instead of space, can be used when you need a specific separator
```

```python
# the following prints one by one. 'cuz default value for end attribute is a new-line character (i.e., \n)
print('hey')
print('there')

# print function with end attribute
print('hey', end=',')     # this line of code prints with an end of comma
print('there')            # this line comes right after the previous line in the same line
```
