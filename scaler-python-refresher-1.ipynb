## Operators

### Arithmatic Operators
Used to perform arithmatic operations
  1. addition (+)
  2. subtraction (-)
  3. multiplication (*)
  4. float division (/)
  4. integer or floor division (//)
  5. modulo (%)
  6. exponentiation (**)

```python
print(24 + 5)
print(24 - 5)
print(24 * 5)
print(24 / 5)   # float division: returns only float eventhough we are performing with ints
print(24 // 5)  # int division: returns the integer just like floor operation in excel
print(-24 // 5) # returns -5 out of the floor operation
print(24 % 5)   # returns the reminder of 4
print(-24 % 5)  # returns 1, 'cuz according to division rules -20 (multiple of 5) is greater than -24
                # so that (-24 - (-25)) ant the reminder would be 1 as mentioned.
                # where -25 is the least multiple of 5 to -24

print(24 ** 4)  # returns the exponential value of 24. i.e., (24 ^ 5)
```

### Comparison Operators

4. greater than or equal to (>=)
5. lesser than or equal to (<=)

### Assignment Operators (=)
For assigning values. i.e., value assignment against variables/objects

The combination of arithmatic operators with this can perform increment and decrement.
*Unfortunately no x++, x--*

```python
x = 1       # assignment

x += 1
print(x)

x *= 2
print(x)

x **= 2
print(x)
```

### Logical Operators
To perform logical operations like making a decision out of more than one conditions.

1. and
2. or
3. not

```python
a = True
b = False
print(a and b)

x = 10
print(x > 1 and x < 12)

s = 'to check'
print(not (s == 'to check'))
```

```python
a, b = 2, 1     # multiple assigments at once
print(a, b)

c = a < b       # integer comparison and returns a bool
print(c)

# implicit type-casting
d = (c < b)     # where, c is a bool
print(d)        # still this works. 'cuz python is a dynamic-type language
                # python performs type-cast automatically for suitable cases

# explicit type-casting
d  = (int(c) < b)
print(d)

# validation
print(int(True), int(False))
```

```python
# can compare strings also, i.e., in a lexicographical order
'abc' > 'xyz'
```

## Control flow
To make conditional statements and for making loops

1. if-else condition
2. while loop
3. for loop

```python
# if-else
age = 15

if age > 60:
  print('eligible and retired')
elif age > 21:
  print('eligible to vote')
else:
  print('not eligible')
```

```python
# one line if-else
eligibility = 'eligible' if age > 21 else 'not eligible'
print(eligibility)
```
