```python
# the following prints one by one. 'cuz default value for end attribute is a new-line character (i.e., \n)
print('hey')
print('there')

# print function with end attribute
print('hey', end=',')     # this line of code prints with an end of comma
print('there')            # this line comes right after the previous line in the same line
```

```python
# variable overwriting
x = 10
print(x, type(x))

# here the data type of a variable has changed by doing this
# this is only possible only because python is a dynamic-type language not like any other static-types such as c++, java
# generally in static-type languages this is not possible at all
```

## Operators

#### Arithmatic Operators
  1. addition:        +
  2. subtraction:     -
  3. multiplication:  *
  4. float division:  /
  4. int division:    //
  5. modulo:          %
  6. exponentiation:  **

```python
print(24 + 5)
print(24 - 5)
print(24 * 5)
print(24 / 5)   # float division: returns only float eventhough we are performing with ints
print(24 // 5)  # int division: returns the integer just like floor operation in excel
print(-24 // 5) # returns -5 out of the floor operation
print(24 % 5)   # returns the reminder of 4
print(-24 % 5)  # returns 1, 'cuz according to division rules -20 (multiple of 5) is greater than -24
                # so that (-24 - (-25)) ant the reminder would be 1 as mentioned.
                # where -25 is the least multiple of 5 to -24

print(24 ** 4)  # returns the exponential value of 24. i.e., (24 ^ 5)
```
