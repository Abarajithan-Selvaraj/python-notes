## Operators

### Arithmatic Operators
Used to perform arithmatic operations
  1. addition (+)
  2. subtraction (-)
  3. multiplication (*)
  4. float division (/)
  4. integer or floor division (//)
  5. modulo (%)
  6. exponentiation (**)

```python
print(24 + 5)
print(24 - 5)
print(24 * 5)
print(24 / 5)   # float division: returns only float eventhough we are performing with ints
print(24 // 5)  # int division: returns the integer just like floor operation in excel
print(-24 // 5) # returns -5 out of the floor operation
print(24 % 5)   # returns the reminder of 4
print(-24 % 5)  # returns 1, 'cuz according to division rules -20 (multiple of 5) is greater than -24
                # so that (-24 - (-25)) ant the reminder would be 1 as mentioned.
                # where -25 is the least multiple of 5 to -24

print(24 ** 4)  # returns the exponential value of 24. i.e., (24 ^ 5)
```

## Control flow
To make conditional statements and for making loops

```python
# if-else
age = 15

if age > 60:
  print('eligible and retired')
elif age > 21:
  print('eligible to vote')
else:
  print('not eligible')
```

```python
# one line if-else
eligibility = 'eligible' if age > 21 else 'not eligible'
print(eligibility)
```

```python
# for loop
for i in range(1, 10, 2) :          # range function: range(start, stop, step)
    print(i, end=', ')              # last index (stop) is exclusive. so it won't print
```

```python
# backward iteration
# 'for i in range(10, 0)' won't work because the compiler doesn't know where to go from the start with +1 as incrementer when the stop is less than the start
for i in range(10, 0, -1):
    print(i, end=', ')
```

## Functions
A block of code with parameters encourange and makes possible of implementation of reusability

```python
# a function to check a number is prime number or not
def isPrime(n):                                 # definition of a functional block
    if n == 2:
        return True
    
    for i in range(2, int(n ** 0.5)):           # optimization for reducing unnecessary runs of the loop
        if n % i == 0:
            return False
    
    return True
```

```python
print(isPrime(21))
print(isPrime(23))
print(isPrime(97))
```

```python
# multiple returns
def math(a, b):
    return a+b, a-b         # this is possible because of 'value packing and unpacking'

add, sub = math(10, 5)      # containing the genies in different jars
print(add, sub)
```

## List
It is a data-structure in which thru we can have a list of elements against a variable  
- list is also a data-type
- not like arrays, list is heterogeneous (i.e., can contain more than one type of data)

```python
lis = []        # empty list, for declaration as well

# initiation
lis = [1, 2, 3, 'a']

print(type(lis))
```

```python
# indexing
print(lis[0])       # returns first element of the list
print(lis[1])       # returns second element of the list
lis[1] = 5          # change second element of the list
print(lis)

# length/size of the list
print(len(lis))

# reverse/negative indexing
print(lis[-1])      # returns the last element of the list, as same as 'lis[len(lis)-1]'
print(lis[-2])      # returns second last element of the list
```

```python
# slicing
lis = [10, 20, 30, 40, 50]
print('full list: ', lis)

# all these slicing creates a copy of the existing list
print('1 to 3: ', lis[1:4])         # get a sublist of the mentioned range
print('until 3: ', lis[:4])         # sublist from the beginning and cutting before 5th element
print('all after 3: ', lis[3:-1])   # sublist from 4th element to the second last using negative indexing
print('using -ve idxing: ', lis[-4: -1])
print('with step: ', lis[::2])      # sublist from start to end with step of 2
print('reversed list: ', lis[::-1]) # reverse the list using step
```

```python
# methods of list object
lis = [10, 20, 30, 40, 50]

# all methods are inplace. i.e., changes only occurs to the list already exists not a new copy
lis.append(60)            # pushbacks a new element
lis.insert(4, 45)         # inserts a new element
lis.sort(reverse=True)    # sort & reverses the list
lis.reverse()             # reverses the list

print(lis)
```

```python
# pre-defined functions
print(sorted(lis, reverse=True))      # sorts the list and creates a new copy
print('orig.: ', lis)
print('sum: ', sum(lis))
print('min: ', min(lis))
print('max: ', max(lis))
```

```python
# list concatenation/entension
print('concat: ', [1,2,3] + [4,5,6])
print('int-multiplication: ', [1,2,3] * 3)
print('list multiplication: ', [1,2,3] * [1,2,3])
```

```python
# interation on list elements
l = [1, 2, 3, 4, 5]
for e in l:
  print(e * 2)        # performing some operations on each element out of the list
```

```python
# copying methods #
a = [1,2,3]

# shallow copy (b is referred to a, so both value and address are same)
b = a

# deep copy (the value stores in a different location after copying)
c = a.copy()

print(b is a)   # 'is' keyword verifies for both the value and address
print(c is a)

print(b == a)   # '==' operator verifies only for value
print(c == a)

# here the proof
print(id(b), ':', id(a))
print(id(c), ':', id(a))
```

```python
# best practice for deep copy
import copy
b = copy.deepcopy(a)

print(b is a)
```

### Nested List

```python
nl = [[1,2,3], [4,5,6], [7,8,9]]

print('0th elem: ', nl[0])
print('(0,0) elem: ', nl[0][0])

# 2nd level slicing
print(nl[2][1:])

# multiplication
print(nl * 2)
```

### List comprehension

```python
# these comprehensions return a new list
# put an equation for all the elements of a list
print([(i**2 + i**3) for i in range(10)])         # equation: (x^2 + x^3)

# having a condition also
print([i**3 for i in range(10) if i%2== 0])       # condition: if a number is even cube the number; only even elems taken

# having more than one conditions
print([i**3 if i%2==0 else i**2 for i in range(10)])    # condition: if num is even cube it, else square it.
```

```python
# left-overs on list
lis = [1, 2, 3, 4, 5]

# single elem change
lis[1] = 10
print(lis)

# multiple elems change at a time (thru slicing)
lis[1:4] = [10, 20, 30]     # it's always safe to have same length
print(lis)

lis[1:4] = [99]             # this is also possible but it's not advisable
print(lis)

# find no. of occurences of an elem
print(lis.count(5))

# find an elem exists in the list or not
print(10 in lis)

# append vs extend
lis.append([1,2,3])         # increases the length of the list by 1
print('append: ',lis)

lis.pop()                   # pops the last elem out of the list
print('after pop: ', lis)

lis.extend([1,2,3])         # increases the length of the list by length of the array to be added
print('extend: ', lis)

print(lis[:3] + [1,2,3])    # as same as extend, but this creates a new list

```
