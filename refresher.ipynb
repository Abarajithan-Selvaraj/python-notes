## Functions
A block of code with parameters encourange and makes possible of implementation of reusability

```python
# a function to check a number is prime number or not
def isPrime(n):                                 # definition of a functional block
    if n == 2:
        return True
    
    for i in range(2, int(n ** 0.5)):           # optimization for reducing unnecessary runs of the loop
        if n % i == 0:
            return False
    
    return True
```

```python
print(isPrime(21))
print(isPrime(23))
print(isPrime(97))
```

```python
# multiple returns
def math(a, b):
    return a+b, a-b         # this is possible because of 'value packing and unpacking'

add, sub = math(10, 5)      # containing the genies in different jars
print(add, sub)
```

## List
It is a data-structure in which thru we can have a list of elements against a variable  
- list is also a data-type
- not like arrays, list is heterogeneous (i.e., can contain more than one type of data)

```python
lis = []        # empty list, for declaration as well

# initiation
lis = [1, 2, 3, 'a']

print(type(lis))
```

```python
# indexing
print(lis[0])       # returns first element of the list
print(lis[1])       # returns second element of the list
lis[1] = 5          # change second element of the list
print(lis)

# length/size of the list
print(len(lis))

# reverse/negative indexing
print(lis[-1])      # returns the last element of the list, as same as 'lis[len(lis)-1]'
print(lis[-2])      # returns second last element of the list
```

```python
# slicing
lis = [10, 20, 30, 40, 50]
print('full list: ', lis)

# all these slicing creates a copy of the existing list
print('1 to 3: ', lis[1:4])         # get a sublist of the mentioned range
print('until 3: ', lis[:4])         # sublist from the beginning and cutting before 5th element
print('all after 3: ', lis[3:-1])   # sublist from 4th element to the second last using negative indexing
print('using -ve idxing: ', lis[-4: -1])
print('with step: ', lis[::2])      # sublist from start to end with step of 2
print('reversed list: ', lis[::-1]) # reverse the list using step
```

```python
# methods of list object
lis = [10, 20, 30, 40, 50]

# all methods are inplace. i.e., changes only occurs to the list already exists not a new copy
lis.append(60)            # pushbacks a new element
lis.insert(4, 45)         # inserts a new element
lis.sort(reverse=True)    # sort & reverses the list
lis.reverse()             # reverses the list

print(lis)
```

```python
# pre-defined functions
print(sorted(lis, reverse=True))      # sorts the list and creates a new copy
print('orig.: ', lis)
print('sum: ', sum(lis))
print('min: ', min(lis))
print('max: ', max(lis))
```

```python
# list concatenation/entension
print('concat: ', [1,2,3] + [4,5,6])
print('int-multiplication: ', [1,2,3] * 3)
print('list multiplication: ', [1,2,3] * [1,2,3])
```

```python
# interation on list elements
l = [1, 2, 3, 4, 5]
for e in l:
  print(e * 2)        # performing some operations on each element out of the list
```

```python
# copying methods #
a = [1,2,3]

# shallow copy (b is referred to a, so both value and address are same)
b = a

# deep copy (the value stores in a different location after copying)
c = a.copy()

print(b is a)   # 'is' keyword verifies for both the value and address
print(c is a)

print(b == a)   # '==' operator verifies only for value
print(c == a)

# here the proof
print(id(b), ':', id(a))
print(id(c), ':', id(a))
```

```python
# best practice for deep copy
import copy
b = copy.deepcopy(a)

print(b is a)
```

### Mutability
Any data structure tend to change its state or content is known as mutability.  
So basically data structures are classified into,  
- Mutable       (list, dict, set)
- Immutable     (rest of the data structurs)

```python
# most of the data-types are immutable
a = 7       # integer
print(a, id(a))

a += 1
print(a, id(a))         # can see different addresses, i.e., integer data-type is immutable
```

### List comprehension

```python
# these comprehensions return a new list
# put an equation for all the elements of a list
print([(i**2 + i**3) for i in range(10)])         # equation: (x^2 + x^3)

# having a condition also
print([i**3 for i in range(10) if i%2== 0])       # condition: if a number is even cube the number; only even elems taken

# having more than one conditions
print([i**3 if i%2==0 else i**2 for i in range(10)])    # condition: if num is even cube it, else square it.
```

```python
# left-overs on list
lis = [1, 2, 3, 4, 5]

# single elem change
lis[1] = 10
print(lis)

# multiple elems change at a time (thru slicing)
lis[1:4] = [10, 20, 30]     # it's always safe to have same length
print(lis)

lis[1:4] = [99]             # this is also possible but it's not advisable
print(lis)

# find no. of occurences of an elem
print(lis.count(5))

# find an elem exists in the list or not
print(10 in lis)

# append vs extend
lis.append([1,2,3])         # increases the length of the list by 1
print('append: ',lis)

lis.pop()                   # pops the last elem out of the list
print('after pop: ', lis)

lis.extend([1,2,3])         # increases the length of the list by length of the array to be added
print('extend: ', lis)

print(lis[:3] + [1,2,3])    # as same as extend, but this creates a new list

```

## Tuples
- Immutable (can't add, append, extend or any other writing operations)
- Read only list
- Memory efficient
- 10x faster than list (due to it's fixed size)
- Syntax: (1, 2, 3) or 1, 2, 3
- Tuples can be heterogenous

```python
tmp = 95, 94, 88, 75
print(type(tmp), tmp)

scores = (95, 94, 88, 75)       # best practice
print(type(scores), scores)
```

```python
# get elems
print('first elem: ', scores[0])
print('last elem: ', scores[-1])

# slicing
print(scores[1:4])

# can't write
scores[1] = 96
```

```python
# methods of tuple object
print('count: ', scores.count(95))
print('index: ', scores.index(95))
```

```python
# tuples are default data structure python uses for returning functions
def foo(a, b):
    return a+b, a-b

x = foo(6, 4)
print(type(x), x)
```

## Dictionary
A data structure of key-value pairs.  
- Unordered (from Python v3.7 dictionaries can be ordered with ordered_dict library, but it's advisable to follow to natural way of dicts)
- Mutable
- keys should be immutable like string, integer, tuple, float, bool (but often used are string and integer)
- values can be any data-type

```python
# initiation
d = dict()                              # empty dictionary
ds = {'k1': 'v1', 'k2': 'v2'}
dlt = {1 : [1, 2, 3], 2 : (1, 2, 3)}

print(type(d), type(ds), type(dlt))
```

```python
# get elems
print(dlt[1])
print(ds['k1'])
print(dlt.get(1))       # the safest way
print(ds.get(0))        # if the key is not found it doesn't return error instead returns None
print(ds.get(0, -1))    # with default value, returns -1 instead of None when the key doesn't exist
print(ds[0])            # throws error since dictionaries are unordered and won't be stacked by index
```

```python
# add/set elems
ds['k3'] = 'v3'                 # add
ds['k2'] = 2                    # set
print(ds)

dlt.update({3 : [7, 8, 9]})     # add
dlt.update({2 : [4, 5, 6]})     # set
print(dlt)
```

```python
# remove elems
ds.pop('k2')
print(ds)
```

```python
# methods of dict object
print(ds.keys())        # returns the list of keys in the dictionary, but the data-type of this list is 'dict_keys'
print(list(ds.keys()))  # casted as list

print(ds.values())      # returns the list of values in the dictionary, but the data-type of this list is 'dict_values'

print(ds.items())       # returns the list of items (in the form of tuples) in the dictionary, but the data-type of this list is 'dict_items'
```

```python
# iteration
print('Using keys method:')
for k in ds.keys():
    print(k, ':', ds[k])

print('\nUsing items method:')
for k, v in ds.items():         # this kind of extraction is tuple packing-unpacking
    print(k, ':', v)
```

#### Tuple packing-unpacking
It's a default functionality in python, helps to handle multiple returns.  
For eg.,

```python
r = foo(5, 6)
print(r, type(r))       # returns a tuple, because of packing

a, s = foo(5, 6)
print(a, s)             # the tuple is unpacked into two different variables
```

*Just like this the above iteration with items method is unpacked.*

```python
# nested dictionary (almost like a json)
nd = {
    'd1' : {
        1 : 1000,
        2 : 2000
    },
    'd2' : {
        3 : 3000,
        4 : 4000,
        5 : 5000,
    },
    'd3' : {
        6 : 6000
    }
}

# find the average of all the values in 'd2'
print('avg: ', sum(nd['d2'].values()) / len(nd['d2'].values()))
```
