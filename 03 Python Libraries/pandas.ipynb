# Day-37

```python
# adding a new row to the dataframe
newRow = {"country":"India",
          "year":"2023",
          "population":"1400000000",
          "continent":"Asia",
          "life_exp":"60",
          "gdp_cap":"2400"}
```

```python
# df.append()   <-- is deprecated and not available with pandas library anymore
```

```python
# method - 2
df.loc[df.shape[0]] = newRow    # loc can add a row, iloc can't..., can use both list and dictionary for assignment
```

```python
df.tail()
```

```python
df.drop(index=[1704],axis=0, inplace=True)    # delete rows using drop
```

```python
df.tail()
```

```python
# to get only duplicate values in a dataframe... (for validation purpose)
df.loc[df.duplicated(subset=["country", "year"])]     # for duplicates found only on the columns mentioned as subset
```

```python
# to remove duplicates
df.drop_duplicates(keep="first", subset=["country", "year"])    # keeps only the first entry
df.drop_duplicates(keep="last")             # keeps only the last entry
df.drop_duplicates(keep=False).tail()       # removes all the entries if duplicates exist
```

```python
# indexing and slicing is possible with both rows and columns together at once with pandas just like numpy
df.loc[3:5, "country":"year"]     # refer lecture notes for more details/combinations of try-outs
```

```python
df.iloc[3:6, :2]     # iloc comes with an advatage of using implicit indexing for columns as well
```

```python
# masking is also possible with pandas just like numpy
df.loc[df["year"]==1952].head()
```
