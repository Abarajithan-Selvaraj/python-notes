```python
import numpy as np
```

```python
# check if all the ids of director column in movies table available in directors table
check = movies["director_id"].isin(directors["id"])
np.all(check)
```

```python
# fetch director's name from directors table and merge it with movies
df = movies.merge(directors, how="inner", left_on="director_id", right_on="id", suffixes=('', '_d'))
df.drop(columns=["director_id", "id_d"], inplace=True)
```

### Missing Values

```python
# means unavailability of data. so, how do they appear?

# usually with numpy or pandas, NaN
# can also be None
```

```python
# how to find them?

dff.isna().sum()          # column wise
dff.isnull().sum()
dff.info()

dff.isna().sum(axis=1).head()    # row wise
```

```python
# basically, in two ways the missing data should be dealt

# 1. Delete the entire row/column
# 2. Impute the missing values based on the characteristics of the data

# This decision is made based on the biz aspects and type/quality of the data
# more than 50% of data is missing, the data-set shouldn't taken for further activities
```

```python
# how to delete those?

dff.dropna()            # deletes the entire row
dff.dropna(how="any")

dff.dropna(how="all")   # deletes only if the values of entire row are NaN

dff.dropna(axis=1).head()      # deletes the entire columns
```

```python
# imputation

dff.fillna(value="0").head()    # fills all the entries where its NaN with 0

dff.fillna(method="ffill")      # forward fill: fills with previous value
dff.fillna(method="bfill")      # backward fill: fills with next value
```

```python
# for advanced imputation

# dff.interpolate()   # need to explore much

# imputation using grouped and aggregate data
dff["Temperature"] = dff["Temperature"].fillna(dff.groupby(by=["Date", "Drug_Name"])["Temperature"].transform(lambda x: x.mean()))

dff["Pressure"] = dff["Pressure"].fillna(dff.groupby(by=["Date", "Drug_Name"])["Pressure"].transform(lambda x: x.mean()))
```

```python
# groupby + multiple aggregation - alternative method
def visitFunc(x):
    x["temp_mean"] = x["Temperature"].mean()
    x["pres_mean"] = x["Pressure"].mean()
    return x

df = dff
df = df.groupby(["Date", "Drug_Name"], group_keys=False).apply(visitFunc)

df
```

```python
# fillup nan values with groupped mean values
df["Temperature"].fillna(df["temp_mean"], inplace=True)
df["Pressure"].fillna(df["pres_mean"], inplace=True)
```
