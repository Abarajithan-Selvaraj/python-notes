```python
movies = pd.read_csv("movies.csv")
directors = pd.read_csv("directors.csv")
```

```python
# initial clean-up: remove unwanted column from the data
movies.drop(columns=["Unnamed: 0"], inplace=True)
directors.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
movies.head()
```

```python
directors.head()
```

```python
import numpy as np
```

```python
# check if all the ids of director column in movies table available in directors table
check = movies["director_id"].isin(directors["id"])
np.all(check)
```

```python
# fetch director's name from directors table and merge it with movies
df = movies.merge(directors, how="inner", left_on="director_id", right_on="id", suffixes=('', '_d'))
df.drop(columns=["director_id", "id_d"], inplace=True)
```

```python
# make this table more readable with money
df["budget"] = (df["budget"]/1000000).round(2)
df["revenue"] = (df["revenue"]/1000000).round(2)
df.rename(columns={"budget": "budget_mil", "revenue": "revenue_mil"}, inplace=True)
```

```python
df.head()
```

```python
# get the highly rated movies i.e., vote_average > 7
df.loc[df["vote_average"] > 7].sort_values(by=["vote_average"], ascending=False).head() # <-- recommended
# df[df["vote_average"] > 7].sort_values(by=["vote_average"], ascending=False).head()   # this works too
```

```python
# top 5 most popular movies
df.sort_values(by=["popularity"], ascending=False).head()
```

```python
# highly rated movies released after 2014
df.loc[(df["vote_average"] > 7) & (df["year"] > 2014)].sort_values(by=["vote_average"], ascending=False).head()
```

```python
# movies released on friday or saturday
df.loc[(df["day"] == "Friday") | (df["day"] == "Saturday")].head()
```

```python
# highest rated movies year wise (just like 'GROUP' in SQL)
df.sort_values(by=["year", "vote_average"], ascending=[False, False]).drop_duplicates(subset=["year"], keep="first").head()
```

### Visit function

```python
def visFunc(x):
    if x == "Male":
        return 1
    
    return 0
```

```python
# visiting a column and idealizing the data
df["gender"].apply(visFunc).head()    # handling with a series
```

```python
# works with more than one columns
def profitFunc(x):
    return x["revenue_mil"] - x["budget_mil"]


df["profit"] = df.apply(profitFunc, axis=1)     # handling with a dataframe
df.head()
```
