```python
movies=pd.read_csv("movies.csv")
movies.drop(columns=["Unnamed: 0"],inplace=True)
movies
```

```python
directors=pd.read_csv("directors.csv")
directors.drop(columns=["Unnamed: 0"],inplace=True)
directors
```

```python
df=movies.merge(directors,how="inner",left_on="director_id",right_on="id",suffixes=('_movie', '_director'))
df.drop(columns=["id_director","director_id"],inplace=True)
df
```

```python
df["budget"]=(df["budget"]/1000000).round(2)
df["revenue"]=(df["revenue"]/1000000).round(2)
df.rename(columns={"budget":"budget_mil","revenue":"revenue_mil"},inplace=True)
df
```

## I want to find out all the risky movies

```python
def custom_func1(x):
    x["risky"]= (x["budget_mil"]-x["revenue_mil"].mean())>=0
    return x
```

```python
df_risky=df.groupby(["director_name"]).apply(custom_func1)
df_risky
```
