```python
df=pd.read_csv("Pfizer_1.csv")
df
```

```python
df.shape
```

```python
df_melt=pd.melt(df,
                id_vars=["Date","Drug_Name","Parameter"],
               var_name="Time",
               value_name="Readings")
df_melt
```

```python
df_pivot=pd.pivot(df_melt,
        index=["Date","Drug_Name","Parameter"],
        columns="Time",
        values="Readings").reset_index()
df_pivot
```

```python
df_pivot.shape
```

```python
df_melt
```

```python
df_final=pd.pivot(df_melt,
        index=["Date","Drug_Name","Time"],
        columns="Parameter",
        values="Readings").reset_index()
df_final
```
